```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Documento = pd.read_csv("/content/sample_data/insurance.csv")
Documento.dropna(axis=0, inplace=True)
print(Documento.info())
reg_lin = LinearRegression()
Documento.sort_values(by=['age'], inplace=True, ascending=True)
x = np.array(Documento['age'])
y = np.array(Documento['charges'])
reg_lin.fit(x.reshape(-1,1), y)
w =  reg_lin.coef_
b = reg_lin.intercept_
print('w = ' + str(reg_lin.coef_) + ', b = ' + str(reg_lin.intercept_))
plt.plot(x,y, "o")
plt.plot(x,y, "o")
plt.plot(x, w*x+b)
plt.grid()
plt.xlabel('Años')
plt.ylabel('Cargo')
plt.show()
```
