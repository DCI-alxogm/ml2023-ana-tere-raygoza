```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Lee los datos desde el archivo con un separador de comas
Documento = pd.read_csv("/content/olympic100m_men.txt", delimiter=",")

# Renombra las columnas
Documento.columns = ['año', 'segundos']

# Calcula la pendiente y la intersección
x = Documento['año']
y = Documento['segundos']
n = len(Documento)

# Calcula la pendiente (m) y la intersección (b)
m = (n * np.sum(x*y) - np.sum(x) * np.sum(y)) / (n*np.sum(x**2) - (np.sum(x))**2)
b = (np.sum(y) - m * np.sum(x)) / n

# Calcula los valores predichos de 'segundos' utilizando la pendiente (m) y la intersección (b)
y_pred = m * x + b

# Crea un gráfico de dispersión (scatter plot) de los datos
plt.scatter(x, y, label='Datos', color='blue', marker='o')

# Dibuja la línea de regresión lineal
plt.plot(x, y_pred, color='red', linewidth=2, label='Regresión Lineal')

# Etiqueta los ejes
plt.xlabel('Año')
plt.ylabel('Segundos')

# Agrega un título al gráfico
plt.title('Gráfico de dispersión con Minimos cuadrados')

# Agrega una leyenda
plt.legend()

# Muestra el gráfico
plt.show()

print("La pendiente (m) es:", m)
print("La intersección (b) es:", b)
```
