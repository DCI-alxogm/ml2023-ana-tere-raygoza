```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def size():
  Documento = open('/content/sample_data/olympic100m_men.txt','r')
  a = sum(1 for line in Documento)
  Documento.close()
  return a

def fun(x,w,b):
    return w*x+b

filas = size()
f = open('/content/sample_data/olympic100m_men.txt','r')
data = np.zeros([filas,2])
counter = 0
for line in f:
  data[counter] = line.split(',')
  counter = counter + 1
f.close()
reg_lin = LinearRegression()
reg_lin.fit(data[:,0].reshape(-1,1), data[:,1])
w =  reg_lin.coef_
b = reg_lin.intercept_
print('w = ' + str(reg_lin.coef_) + ', b = ' + str(reg_lin.intercept_))
plt.plot(data[:,0],data[:,1], "o")
plt.plot(data[:,0],data[:,1], "o")
plt.plot(data[:,0],[fun(i,w,b) for i in data[:,0]] )
plt.grid()
plt.xlabel('AÃ±os')
plt.ylabel('Segundos')
plt.show()
```
